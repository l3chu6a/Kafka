```python
from atproto import Client

# Paso 1: Autenticarse
client = Client()
client.login('l3chu6a.bsky.social', 'Bluesky_005')

# Paso 2: Obtener timeline (últimos posteos)
timeline = client.app.bsky.feed.get_timeline({'limit': 100})

# Paso 3: Procesar y mostrar los posteos
for item in timeline.feed:
    post = item.post
    print("Autor:", post.author.handle)
    print("Texto:", post.record.text)
    print("-" * 40)

print(f"\nTotal de posteos obtenidos: {len(timeline.feed)}")
```
